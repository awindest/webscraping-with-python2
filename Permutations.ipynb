```python
# A Python program to print all combinations  
# with an element-to-itself combination  
from itertools import combinations_with_replacement  
  
# Get all combinations of ["Y", "N", ""] and length 10 
comb = combinations_with_replacement(["Y", "N", "","N","Y"], 10)  
count = 0 
# Print the obtained combinations  
for i in list(comb):  
    print (i)
    count += 1
goodcount = count
count = 0
comb2 = combinations_with_replacement(["X", "N", "","N","Y"], 10)  
for i in list(comb2):  
    print (i)
    count += 1
print(goodcount-1, count-1)
```
